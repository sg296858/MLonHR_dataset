```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['MarriedID','MaritalStatusID','PositionID','Position','HispanicLatino','TermReason','EmploymentStatus','PerformanceScore','EmpSatisfaction','SpecialProjectsCount','Absences']]
Y=df['Salary']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.tree  import DecisionTreeRegressor
model=DecisionTreeRegressor(criterion='squared_error', splitter='best', max_depth=4)
model.fit(X_train,Y_train)
```

```python
Y_test
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
r2_score(y_pred,Y_test)
```
